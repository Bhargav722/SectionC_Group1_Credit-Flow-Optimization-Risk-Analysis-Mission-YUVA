# Step 05: Final Load & Preparation
Goal: Finalize column naming and data types, and save the polished dataset for downstream use.

```python
import pandas as pd
import os

INPUT_FILE = "../data/interim/02_cleaning.csv"
OUTPUT_FILE = "../data/processed/cleaned_mission_yuva.csv"

# Ensure output directory exists
os.makedirs("../data/processed", exist_ok=True)

df = pd.read_csv(INPUT_FILE)
print("✅ Loaded cleaned dataset for final prep")
```

```python
# Rename Columns to Snake Case
df = df.rename(columns={
    'Application ID'            : 'application_id',
    'District Name'              : 'district_name',
    'Gender'                     : 'gender',
    'Residential Type'           : 'residential_type',
    'Specially Abled Type'       : 'specially_abled_type',
    'Enterprise Type'            : 'enterprise_type',
    'Sector'                     : 'sector',
    'IFSC Code'                  : 'ifsc_code',
    'Days_to_BHD_Verification'   : 'days_to_bhd_verification',
    'Days_to_SBDU_Verification'  : 'days_to_sbdu_verification',
    'Days_to_DLIC_Approval'      : 'days_to_dlic_approval',
    'Days_to_Bank_Sanction'      : 'days_to_bank_sanction',
    'Days_to_Disbursement'       : 'days_to_disbursement',
    'Total_TAT_Days'             : 'total_tat_days',
    'Loan_Amount_INR'            : 'loan_amount_inr',
    'Project_Cost_INR'           : 'project_cost_inr',
    'Applicant_Monthly_Income'   : 'applicant_monthly_income',
    'Loan_Tenure_Months'         : 'loan_tenure_months',
    'Interest_Rate_%'            : 'interest_rate',
    'EMI_Amount'                 : 'emi_amount',
    'EMI_to_Income_Ratio'        : 'emi_to_income_ratio',
    'Risk_Score'                 : 'risk_score',
    'Credit_Eligibility_Score'   : 'credit_eligibility_score',
    'Repayment_Status'           : 'repayment_status'
})

print("✅ Columns renamed to snake_case")
```

```python
# Fix Data Types & Rounding
df['application_id'] = df['application_id'].astype(int)
df['district_name'] = df['district_name'].str.title()

float_cols = df.select_dtypes(include=['float']).columns
df[float_cols] = df[float_cols].round(2)

print("✅ Data types fixed and rounded")
```

```python
df.to_csv(OUTPUT_FILE, index=False)
# Also save to the legacy 'clean' folder for backward compatibility if needed
os.makedirs("../clean", exist_ok=True)
df.to_csv("../clean/cleaned_mission_yuva.csv", index=False)

print(f"✅ Final dataset saved to {OUTPUT_FILE} and ../clean/cleaned_mission_yuva.csv")
print(f"Final shape: {df.shape}")
```
